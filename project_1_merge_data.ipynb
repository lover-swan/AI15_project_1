```python
import pandas as pd
```

```python
aisles=pd.read_csv('/content/drive/MyDrive/project_1/aisles.csv')
departments=pd.read_csv('/content/drive/MyDrive/project_1/departments.csv')
order_prior=pd.read_csv('/content/drive/MyDrive/project_1/order_products__prior.csv')
order_train=pd.read_csv('/content/drive/MyDrive/project_1/order_products__train.csv')
orders=pd.read_csv('/content/drive/MyDrive/project_1/orders.csv')
products=pd.read_csv('/content/drive/MyDrive/project_1/products.csv')
```

```python
#제품 정보 데이터 병합 
```

```python
products=pd.merge(products,aisles,on='aisle_id',how='outer')
products=pd.merge(products,departments,on='department_id',how='outer')
```

```python
products=products.drop(columns=['aisle_id','department_id'])
```

```python
products.to_csv(path_or_buf='/content/drive/MyDrive/project_1/after/products.csv',index=False)
```

```python
#주문정보 데이터 병합 
```

```python
order=pd.concat([order_prior,order_train])
```

```python
order=order.reset_index(drop=True)
```

```python
order
```

```python
orders=orders.drop(columns=['order_hour_of_day','eval_set','order_number','order_dow'])
orders
```

```python
orders=pd.merge(order,orders,on='order_id',how='outer').dropna().reset_index(drop=True)
```

```python
orders.to_csv(path_or_buf='/content/drive/MyDrive/project_1/after/orders.csv',index=False)
```
